# Conteo

```python
from collections import Counter
cnt =  Counter()
s = "Hola mundo hola todos"
dat={}
for c in s:
    if c in dat:
        dat[c] = dat[c] + 1
    else:
        dat[c] = 1

            
```

```python
print(dat)
```

```python
print(list(dat.keys()))
print(list(dat.values()))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

k=list(dat.keys())
v=list(dat.values())
print(k,v)
x_pos = [i for i, _ in enumerate(k)]
plt.xticks(x_pos, k)
plt.bar(x_pos, v, align='center')
```

# Similitud

# Coeficiente de Jaccard 
# $$J(A,B) = \frac{|A \cap B|}{|A \cup  B|}$$

```python
a="casuchas"
b="casitas"
```

```python
CA=set(a)
CB=set(b)
tI=len(CA.intersection(CB))
tU=len(CA.union(CB))
J=tI/tU
```

```python
J
```

# Coeficiente de Dice 
# $$D(A,B) =  \frac{ 2 \times |A \cap B|}{|A| + |B|}$$

```python
D=2*tI/(len(CA) + len(CB))
```

```python
D
```

# preprocesamiento de los datos

```python
# remover lineas vacias
text2=[]
with open("celestina.txt", 'r', encoding='utf8') as f:
    for l in f.readlines():
        if not l.strip():
            continue
        if l.strip().startswith('#'):
                continue
        text2.append(l.strip().split())
    f.close()
```

```python
text2[:3]
```


# Conteo de términos

```python
from collections import Counter
tt=[]
for t in text2:
    for w in t:
        for c in w:
            tt.extend(c.lower())
            
di=Counter(tt)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

k=list(di.keys())
v=list(di.values())
x_pos = [i for i, _ in enumerate(k)]
plt.xticks(x_pos, k)
plt.bar(x_pos, v, align='center')
```
